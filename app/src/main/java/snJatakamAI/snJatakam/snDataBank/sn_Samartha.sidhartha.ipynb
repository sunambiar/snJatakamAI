```python
#!pip3 install multipledispatch 
from multipledispatch import dispatch 

class HMS:
    
  hrs = float(0.0) 
  
  #constructor
  def __init__(self, h) :
    self.hrs = h 

  @staticmethod
  @dispatch(float)
  def hms(h) :
    return HMS.hms(h, False)

  @staticmethod
  @dispatch(float, bool)
  def hms(h, hr24) :
    sgn = ''
    if not hr24 :
      if h > 24.0 :
        h -= 24 * (int(h / 24))
      if h < 0 :
        sgn = '-'
        h = -h
      hrs = h
      hh = int(h)
      h = (h - int(h)) * 60
      mm = int(h)
      h = (h - int(h)) * 60
      ss = round(h)
      if ss >= 60 :
        ss -= 60
        mm += 1
      if mm >= 60 :
        mm -= 60
        hh += 1
      if hr24 :
        ampm = 'hrs'
      else :
        if hh >= 24 :
          hh -= 24
        if hh > 12 :
          hh -= 12
          ampm = 'PM'
        else :
          if (hh + (mm + ss / 60.0) / 60.0) > 12.0 :
            ampm = 'PM'
          else :
            ampm = 'AM'
    strfmt = "{}{}:{}:{} {}"
    return (strfmt.format(sgn.strip(), hh, mm, ss, ampm)) 



if __name__ == "__main__":
  print(HMS.hms(26.54234))

```

```python
"""
/*--------------------------------------------------------*/
/*               Suresh's  Jatakam - classes              */
/*  Python code : 2020-11-30                              */
/*--------------------------------------------------------*/
package snMobileJatak.sn_Jatak.snClasses.snBasic;
//package snBasic;
/*--------------------------------------------------------*/
"""
#!pip3 install multipledispatch 
from multipledispatch import dispatch  

class snDate :

  CC, YY, yy, mm, dd, hh, mt, ss = 0, 0, 0, 0, 0, 0, 0, 0

  
  #constructor
  @dispatch(int, int, int, int, int, int)
  def __init__(self, y1, m1, d1, h1, mt1, s1) :
    yy = y1
    YY = yy % 100
    self.yy = y1
    self.mm = m1
    self.dd = d1
    self.hh = h1
    self.mt = mt1
    self.ss = s1
    self.YY = YY
    self.CC = (yy - YY)

  @classmethod
  @dispatch(int, int, int, int, int, int)
  def snDate(cls, y1, m1, d1, h1, mt1, s1) :
    cls.yy = y1
    cls.mm = m1
    cls.dd = d1
    cls.hh = h1
    cls.mt = mt1
    cls.ss = s1
    cls.YY = cls.yy % 100
    cls.CC = cls.yy - cls.YY

  @classmethod
  @dispatch(int, int, int)
  def snDate(cls, y1, m1, d1) :
    cls.yy = y1;
    cls.mm = m1;
    cls.dd = d1;
    cls.YY = cls.yy % 100;
    cls.CC = cls.yy - cls.YY;


  @classmethod
  #@dispatch(class)
  def snDate(cls, d) :
    cls.yy = d.yy;
    cls.mm = d.mm;
    cls.dd = d.dd; 
    cls.YY = cls.yy % 100;
    cls.CC = cls.yy - cls.YY;


  @classmethod
  @dispatch(int)
  def snDate(cls, jd) :
    #long a, l, y, m, d;
    y = int(jd / 365.25)
    a = y / 100;
    jd -= (365 * y + y / 4 + 2 - a + a / 4)
    m = int((jd / 30.6) - 1)
    l = int(30.6 * (float(m + 1)))
    jd -= l
    d = jd
    cls.yy = y
    cls.mm = m
    cls.dd = d
    cls.YY = (cls.yy % 100)
    cls.CC = (cls.yy - cls.YY)

  @classmethod
  @dispatch(float)
  def snDate(cls, JD) :
    sD = cls.JD2CAL(JD)
    cls.yy = sD.yy
    cls.mm = sD.mm
    cls.dd = sD.dd
    cls.YY = sD.YY
    cls.CC = sD.CC
    cls.hh = sD.hh
    cls.mt = sD.mt
    cls.ss = sD.ss

  @classmethod
  def dmy(cls) :
    return ('{}/{}/{}'.format(cls.dd, cls.mm, cls.yy))

  @classmethod
  def ymd(cls) :
    return ('{}/{}/{}'.format("{} Yrs {} mths {} days", cls.yy, cls.mm, cls.dd))


  @classmethod
  def years(cls) :
    return (float(cls.yy) + (float(cls.mm) + float(cls.dd) / 30.0) / 12.0)

  @staticmethod
  @dispatch(int, int, int)
  def jd(d, m, y) :
    #long a, j, l;
    #float b;
    if (m < 3) :
      m += 12
      y -= 1
    a = int(y / 100);
    b = 30.6 * (float(m + 1))
    l = int(b);
    j = int(365 * y + y / 4 + l + 2 - a + a / 4 + d)
    return (j)

  @staticmethod
  #@dispatch(snDate, snDate)
  def datediff(d1, d2) :
    '''
    /*
    long ld1, ld2;
    ld1 = jd(d1.dd, d1.mm, d1.yy);
    ld2 = jd(d2.dd, d2.mm, d2.yy);
     */'''
    #double ld1, ld2;
    ld1 = float(JD(d1.yy, d1.mm, d1.dd, 12, 0, 0))
    ld2 = float(JD(d2.yy, d2.mm, d2.dd, 12, 0, 0))
    return int(ld1 - ld2)


  @staticmethod
  #@dispatch(snDate, snDate)
  def dateDiff(d1, d2) :
    x = snDate(d1)
    #long ld, xx;
    ld = datediff(d1, d2);
    #//System.out.println(" Date Difference :--------->"+d1.dmy() +" "+ d2.dmy());
    x.yy = int(ld / 365)
    #//ld = Math.abs(ld);
    xx = (abs(ld) - abs(x.yy) * 365)
    x.mm = int(xx / 30)
    x.dd = int(xx - x.mm * 30)
    if x.dd >= 30 :
      x.dd = 0
      x.mm += 1
    if x.mm >= 12 :
      x.mm = 0
      x.yy += 1 
    #//System.out.println(x.dmy());
    return (x)


  @staticmethod
  #@dispatch(snDate, int)
  def dateAdd(d1, dadd) :
    '''
    /*
    double ld1;
    ld1 = JD(d1.yy, d1.mm, d1.dd,12,0,0);
    ld1 = ld1 + dadd;
    return (new snDate(ld1));
     */
     '''
    return (snDate(dadd + snDate.JD(d1.yy, d1.mm, d1.dd, 12, 0, 0)))


  @staticmethod
  @dispatch(float)
  def Fix(x) :
   return (int(x))



  @staticmethod
  @dispatch(int, int, int, int, int, float, float)
  def day2000(year, month, day, hour,
         min, sec, greg) :
    #// Returns days before J 2000.0 given date in Gregorian calendar
    #// (greg = 1) or Julian calendar (greg = 0)
    #double a;
    #int b;
    a = 10000.0 * year + 100.0 * month + day;
    if (month <= 2) :
      month += 12;
      year -= 1;
    if (greg == 0) :
      b = -2 + Fix((year + 4716) / 4) - 1179
    else :
      b = Fix(year / 400) - Fix(year / 100) + Fix(year / 4)
    a = 365.0 * year - 730548.5;
    return float(a + b + Fix(30.6001 * (month + 1)) + day +
        (hour + month / 60.0 + sec / 3600.0) / 24.0)


  @staticmethod
  @dispatch(int, int, int, int, int, float)
  def JD(y, m, d, h, mn, s) :
    #// Provide y = negative y  if Era is Before Common Era (BCE)
    #long jy, ja, jm;    //scratch

    era = "CE"      # Common Era

    if (y == 0) :
      print("There is no year 0 in the Julian system!")
      return -1
    else :
      if (y < 0) :
        era = "BCE"    # Before Common Era
        y = -y
    
    if (y == 1582 and m == 10 and d > 4 and d < 15) :
      print("The dates 5 through 14 October, 1582, do not exist in the Gregorian system!")
      return -2
    
    # //  if( y < 0 )  ++y;
    if (era == "BCE") :
      y = -y + 1
    
    if (m > 2) :
      jy = y
      jm = m + 1
    else :
      jy = y - 1
      jm = m + 13

    intgr = int((((365.25 * jy) // 1) + ((30.6001 * jm) // 1) + d + 1720995) // 1)

    #//check for switch to Gregorian calendar
    gregcal = 15 + 31 * (10 + 12 * 1582)
    if (d + 31 * (m + 12 * y) >= gregcal) : 
      ja = int((0.01 * jy) // 1)
      intgr += 2 - ja + int((0.25 * ja) // 1) 
  

    #//correct for half-day offset
    dayfrac = float(h) / 24.0 - 0.5
    if (dayfrac < 0.0) :
      dayfrac += 1.0
      intgr -= 1
    

    #//now set the fraction of a day
    frac = dayfrac + (float(mn) + float(s) / 60.0) / 60.0 / 24.0;

    #//round to nearest second
    jd0 = (float(intgr) + frac) * 100000
    jd = int(jd0 // 1)
    if (jd0 - jd > 0.5) :
      jd += 1
    
    return float(jd / 100000) 



  '''
  //-----------------------------------------------------------------------------
  // convert Julian date to calendar date
  // (algorithm adopted from Press et al.)
  //-----------------------------------------------------------------------------
  '''
  @staticmethod
  @dispatch(float)
  def JD2CAL(jd) :
    #long j1, j2, j3, j4, j5;      //scratch

    #//
    #// get the date from the Julian day number
    #//
    intgr = int(jd // 1)
    frac = jd - intgr
    gregjd = 2299161
    if (intgr >= gregjd) :      # //Gregorian calendar correction
      tmp = int((((intgr - 1867216) - 0.25) // 1) / 36524.25)
      j1 = intgr + 1 + tmp - int((0.25 * tmp) // 1)
    else :
      j1 = intgr
  
    #//correction for half day offset
    dayfrac = frac + 0.5
    if (dayfrac >= 1.0) :
      dayfrac -= 1.0
      j1 += 1

    j2 = j1 + 1524
    j3 = int((6680.0 + ((j2 - 2439870) - 122.1) / 365.25) // 1)
    j4 = int((j3 * 365.25) // 1)
    j5 = int(((j2 - j4) / 30.6001) // 1)

    d = int((j2 - j4 - ((j5 * 30.6001) // 1)) // 1)
    m = int((j5 - 1) // 1)
    if (m > 12) :
      m -= 12
    
    y = int((j3 - 4715) // 1)
    if (m > 2) :
      y -= 1
    
    if (y <= 0) :
      y -= 1

    #//
    #// get time of day from day fraction
    #//
    hr = int((dayfrac * 24.0) // 1)
    mn = int(((dayfrac * 24.0 - hr) * 60.0) // 1)
    f = ((dayfrac * 24.0 - hr) * 60.0 - mn) * 60.0
    sc = (f // 1)
    f -= sc
    if (f > 0.5) :
      sc += 1

    return (snDate(y, m, d, hr, mn, sc))



```
